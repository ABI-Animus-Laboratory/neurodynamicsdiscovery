```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/hpc/mzhu843/modelling/nest/DATA/raw/experiment3_unfiltered_raw.csv')
del df[df. columns[0]]
```

```python
np_array = df.values
```

```python

# Assuming 'arr' is your NumPy array
first_row = np_array[6, :]

# Get indices of non-zero elements in the first row
non_zero_indices = np.nonzero(first_row)[0]

# Create a dictionary from indices and their corresponding values
non_zero_dict = {index: first_row[index] for index in non_zero_indices}

# non_zero_dict now contains the indices and non-zero elements of the first row
print(non_zero_dict)
```

```python
for i in range(np.size(np_array, axis=0)):
    counter = 0
    for j in range(np.size(np_array, axis=1)):
        if np_array[i][j] != 0 and counter == 0:
            np_array[i][j] = 1
            counter = 2
        elif np_array[i][j] != 0 and counter != 0:
            np_array[i][j] = 0
        elif np_array[i][j] == 0 and counter != 0:
            counter -= 1
```
