```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
file = np.load('/hpc/mzhu843/modelling/nest/DATA/processed/spikes/experiment3_unfiltered_processed.npy')
spike_numbers = np.sum(file, axis=1)
bins = range(1,300,10)
plt.figure()
plt.hist(spike_numbers, bins, rwidth=0.8)
plt.xlabel('Number of Spikes')
plt.ylabel('Number of Neurons')
plt.title('Neuron densities for total amount of spiking')
plt.savefig('/hpc/mzhu843/modelling/nest/results/neuron_densities_for_spike_frequencies.png')

```
